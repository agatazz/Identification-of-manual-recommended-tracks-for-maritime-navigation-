```python
import geopandas as gdp

found_corridors = gdp.read_file("../data/processed/final.gpkg")

# adjust weights according to business logic and preferences

CONSENSUS_SCORE_WEIGHT = 0.4
UNIQE_VESSEL_COUNT_WEIGHT = 0.3
TIDE_INDEPENDENCE_RATION = 0.3
```

```python
def calculate_Path_Reliability_Score(found_tracks):
    weights = {
        'consensus_score': CONSENSUS_SCORE_WEIGHT,
        'unique_vessel_count': UNIQE_VESSEL_COUNT_WEIGHT,
        'tide_independent_ratio': TIDE_INDEPENDENCE_RATION
    }
    
    # normalization
    found_tracks['consensus_score_norm'] = found_tracks['consensus_score'] / found_tracks['consensus_score'].max()
    found_tracks['unique_vessel_count_norm'] = found_tracks['unique_vessel_count'] / found_tracks['unique_vessel_count'].max()
    found_tracks['tide_independent_ratio_norm'] = found_tracks['tide_independent_ratio']  # Already between 0 and 1
    
    found_tracks['path_reliability_score'] = (
        weights['consensus_score'] * found_tracks['consensus_score_norm'] +
        weights['unique_vessel_count'] * found_tracks['unique_vessel_count_norm'] +
        weights['tide_independent_ratio'] * found_tracks['tide_independent_ratio_norm']
    )
    
```

```python
prs_calculated = calculate_Path_Reliability_Score(found_tracks)
```

```python
prs_calculated.to_file("ranked_trajecotries.gpkg", driver="GPKG")
```
