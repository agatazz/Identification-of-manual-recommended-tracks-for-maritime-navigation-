Prepare s-57 water data for the area study. Get all water polygons which are shallower then 3 m

```python
import geopandas as gpd

DEPTH = 3
depth_gdf = gpd.read_file("..data/processed/water_data.gpkg")
```

```python
shallow = depth_gdf[depth_gdf["DRVAL2"] <= DEPTH]
```

```python
shallow.to_file("../data/processed/shallow_waters.gpkg", driver="GPKG")
```
